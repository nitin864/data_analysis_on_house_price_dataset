```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
#replacing blanks with 0 as tenure is 0 and no total charges are recorded
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df["customerID"].duplicated().sum()
```

```python
#converting "SeniorCitizen" Binary Value to "yes" or "no"
```

```python
def convert(value):
  if value == 1:
    return "yes"
  else:
    return "no"
df["SeniorCitizen"] = df["SeniorCitizen"].apply(convert)
```

```python
#---------------------------------------------------------DATA_INSPECTION_DONE-----------------------------------------------------------------------------------------------------------------------------
```

```python
x = sns.countplot(x = 'Churn' , data = df)
x.bar_label(x.containers[0])
plt.title('Customer Count by Churn')
plt.show()
```

```python
#Visualization 1 : from the given chart we conclude that 1869 customer churned out
```

```python
plt.figure(figsize = (5,5))
g = df.groupby('Churn').agg({'Churn': 'count'})
plt.pie(g['Churn'], labels = g.index, autopct = '%1.2f%%')
plt.title('Customer % by Churn')
plt.show()
```

```python
#Visualization 2 : from the given pie chart we conclude that 26% customer churned out
```
